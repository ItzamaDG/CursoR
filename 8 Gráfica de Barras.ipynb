```r
#Usando R base
#Necesita una tabla donde este la variable categorica
table(mtcars$cyl)
#Aqui divide por clases y me da cuantos hay de cada uno
```

```r
barplot(table(mtcars$cyl),
       col = 'green',
       border = 'red',
       main = 'titulo',
       sub = 'subtitulo',
       xlab = 'Etiqueta X',
       ylab = 'Etiqueta Y',
       horiz = TRUE)
#border se refiere a los de la barra
```

```r
#Ahora hagamoslo con ggplot2
library(ggplot2)
ggplot(data = mtcars,mapping=aes(x = factor(cyl)))+
geom_bar()+coord_flip()
```

```r
#Variables separadas
p<-ggplot(data = mtcars,
         mapping = aes(x = factor(cyl),
                      fill = factor(gear)))
```

```r
p+geom_bar(position = 'stack',
          stat = 'count')
#Honestamente sin stack se ve igual
```

```r
p+geom_bar(position = 'dodge',stat = 'count')
```

```r
p+geom_bar(position='fill',stat='count' )
```
