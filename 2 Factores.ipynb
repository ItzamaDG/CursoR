```r
#Ya sabes que es una variable categorica
tallas <-c('m','g','S','S','m','M')
#Creamos un factor a traves de este vector
tallas_factor <- factor(tallas)
plot(tallas_factor)
```

```r
#Cambiemos los bombres de los niveles
tallas_refactor <- factor(tallas,
                          levels = c('g','m','M','S'),
                         labels = c('G','M','M','S'))
#Renombramos  todo a mayuscula
plot(tallas_refactor)
```

```r
#La grafica no esta en orden, hay que darle un orden
tallas_ordenado<- factor(tallas, ordered = TRUE,
                          levels = c('S','m','M','g'),
                         labels = c('S','M','M','G'))
plot(tallas_ordenado)
```
