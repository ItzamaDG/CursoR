```r
library(dplyr)
```

```r
filter(gapminder,continent=='Asia')
```

```r
#Usamos el operador pipe
gapminder %>% filter(continent == 'Asia')
```

```r
#Mas de una condicion
gapminder%>%filter(lifeExp<=40,year == 2002)
```

```r
gapminder %>% filter(continent == 'Asia', year == 2007)%>%summarise(conteo = n())
#Cuenta cuantos
```

```r
#El maximo por ejemplo
gapminder %>% filter(continent == 'Asia', year == 2007)%>%
summarise(MaxEsp = max(lifeExp))
#O sea primero filtra los paises que satisfacen eso y despues pidele algo, una media
#de alguna columna o que se yo
```

```r
#groupby
gapminder%>%group_by(year)%>%summarise(MaxEsp=max(lifeExp))
#Es decir, para cada año, toma el maximo de la esperanza de vida
```
