```r
year <- c('2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018')
disney <- c(11, 13, 11, 8, 12, 11, 12, 8, 10)
plot(year,disney)
```

```r
#GGPLOT
#Se ve por capas
install.packages('ggplot2')
library(ggplot2)
```

```r
peliculas <- data.frame(year,disney)
```

```r
ggplot(data = peliculas,mapping = aes(x = year,
                                     y = disney)) +
      geom_point() +labs(title='disney')
#mapping es quÃ© dato va en donde
#la capa geom cambia segun el tipo de grafica
```
