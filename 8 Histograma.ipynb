```r
#Frecuencia por intervalos de una variable
library(ggplot2)
```

```r
#Llamamos a los datos precargados en R
data('mtcars')
```

```r
#Con RBase
#Funcion hist, recibe un vector
#hist(mtcars$hp,breaks = 50)
hist(mtcars$hp,breaks = seq(50,350,50),
    col = 'darkgray',
    border = 'gray10',
    main = 'Mi Histograma',
    xlab = 'Label Eje X',
    ylab = 'Conteo')
#Modificamos los bins (La cantidad de intervalos)
#De 50 a 350 de 50 en 50
```

```r
#Ahora hagamoslo usango ggplot 2
ggplot(data = mtcars,
      mapping=aes(x=hp))+
      geom_histogram(bins = 9) 
```

```r
#Ahora hagamos esto pero diferenciando por tipo de motor
#Es decir, quiero que se se pinten las barras
#Segun el color del motor al que pertenecen+
#Usamos el factor porque vs es por defecto de tipo
#Caracter
ggplot(data = mtcars,
      mapping=aes(x=hp,fill = factor(vs)))+

      geom_histogram(bins = 9,
                    position = 'identity',
                    alpha = 0.8) +
      labs(title = 'Histograma',
          x = 'HorsePower',
          y = 'Conteo',
          subtitle = 'subtitulo',
          caption = 'Breve descripci√≥n')
```
