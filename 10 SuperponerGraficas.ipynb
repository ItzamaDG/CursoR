```r
library(ggplot2)
```

```r
ggplot(data = iris,
      mapping = aes(x=Sepal.Length,y=Sepal.Width,color=Species))+ geom_point()+
geom_smooth(method='lm')
#Me colorea por especies
#Superponemos una geometria que me de algo como una tendencia
#lm significa que es de forma lineal
```

```r
#Ahroa de lineas
#USamos 50 porque justo son 50 observaciones de flores setosa
ggplot(data=iris, 
       mapping = aes(x = rep(1:50,3), 
        y = Petal.Width,color = Species))+
    geom_line()
#El rep es porque so n3 especies
```
